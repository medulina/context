```python
import numpy as np
from scipy import ndimage
%pylab inline
```

```python
co = ndimage.imread('DentateGyrusSurface_black.png')
```

```python
imshow(co)
```

```python
for i in range(0,169):
    coi = 672*i//169
    coj = 672*(i+1)//169
    con = co.copy()
    con[:,coi:coj,:3]=np.array([[[255,246,0]] * (coj-coi)]*con.shape[0])
    imsave('slice%03d.jpg'%i,con, format = 'jpg')
```
